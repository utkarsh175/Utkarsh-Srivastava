```python
def purchase(df):
    if(df[(df['ONEOFF_PURCHASES']==0) & (df['INSTALLMENTS_PURCHASES']==0)].shape):
        return 'none'
    if(df[(df['ONEOFF_PURCHASES']>0) & (df['INSTALLMENTS_PURCHASES']>0)].shape):
        return 'both_one_off'
    if(df[(df['ONEOFF_PURCHASES']>0) & (df['INSTALLMENTS_PURCHASES']==0)].shape):
        return 'one_one_off'
    if(df[(df['ONEOFF_PURCHASES']==0) & (df['INSTALLMENTS_PURCHASES']>0)].shape):
        return 'one_installment_purchase'
```

```python
#Clustering 
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters =6, random_state =123)
kmeans.fit(reduced_df)
```

```python
pd.Series(kmeans.labels_).value_counts()
```

```python
wcss = []
cluster_list = range(1,6)
for i in cluster_list :
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 40)
    kmeans.fit(reduced_df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(cluster_list, wcss)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

The number of clusters as defined by the elbow method are likely to be 3
